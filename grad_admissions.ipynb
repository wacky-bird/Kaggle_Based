```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import sys
import os
```

```python
#Read in the dataset
df = pd.read_csv("Admission_Predict.csv",sep = ",")
```

```python
#Check to see the number of columns in the dataframe and print them out
print("Checking the number of columns...",len(df.columns),"columns" )
for x in df.columns:
    print(str(x)+'...')
```

LOR and Chance of Admit have a space character after the coumn name, a rename is required to clean it up

```python
#Renaming certains columns for ease
df = df.rename(columns = {'LOR ':'LOR',
                          'Chance of Admit ':'Chance of Admit'})
```

Let's use seaborn to create a heat map because I like the way it looks and it works well to show off this dataset.

```python
#Create a heatmap using matplotlib and seaborn
fig,ax = plt.subplots(figsize=(10,10))
sns.heatmap(df.corr(), ax=ax, annot=True, 
            linewidth=0.05, fmt='.2f', cmap='RdBu')
```
